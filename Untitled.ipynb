```python
def a(record):
    ze = np.zeros((500,))
    record=re.sub('[^ㄱ-힣]', ' ',record).split()

    for key, value in Counter(record).items():
        if key in vocab:
            ind = vocab.index(key)
            ze[ind]=1+np.log(value)
    return ze
```
